```python
import requests
import pandas as pd

response = requests.get(
    "https://jsonplaceholder.typicode.com/users"
)
data = response.json()

df_pd = pd.json_normalize(data)
df_spark = spark.createDataFrame(df_pd)

from pyspark.sql.functions import col

df_selected = df_spark.select(
    col("name"),
    col("username"),
    col("`address.city`").alias("city"),
    col("`company.name`").alias("company_name")
)
display(df_selected)


# Convert the parse data into list of dictionary and write it to new csv 
dict_list = [row.asDict() for row in df_selected.collect()]
import csv

with open('api_users.csv', 'w', newline='') as file:
    writer = csv.DictWriter(file, fieldnames=dict_list[0].keys())
    writer.writeheader()
    writer.writerows(dict_list)

# define a function and handle api failure (non-200 responses) using try-except 

def fetch_and_save_api_users(api_url, output_csv):
    import requests
    import pandas as pd
    import csv

    try:
        response = requests.get(api_url)
        if response.status_code != 200:
            print(f"API request failed with status code {response.status_code}")
            return

        data = response.json()
        df_pd = pd.json_normalize(data)
        df_spark = spark.createDataFrame(df_pd)

        from pyspark.sql.functions import col

        df_selected = df_spark.select(
            col("name"),
            col("username"),
            col("`address.city`").alias("city"),
            col("`company.name`").alias("company_name")
        )
        display(df_selected)

        dict_list = [row.asDict() for row in df_selected.collect()]
        with open(output_csv, 'w', newline='') as file:
            writer = csv.DictWriter(file, fieldnames=dict_list[0].keys())
            writer.writeheader()
            writer.writerows(dict_list)
    except Exception as e:
        print(f"Error: {e}")

fetch_and_save_api_users("https://jsonplaceholder.typicode.com/users", "api_users.csv")
```
