```python
def calculate_bonus(salary):
    if salary > 40000:
        return salary + 5000
    elif salary > 30000:
        return salary + 3000 
    elif salary > 20000:
        return salary + 2000
    else:
        return 'No Bonus' 
        

calculate_bonus(35000)    
```

```python
# IF- ELSE Conditional statements 

Salary = int(input('Enter the amount of salary you earn.'))
if Salary > 70000:
    print('Great Salary! High Class')
elif Salary > 50000:
    print('Good Salary! Middle Class')
elif Salary > 30000:
    print('Average Salary! Working Class')
else:
    print('Poor Salary! Lower Class')
```

```python
Customers = {'name': ['Rahul','Devraj', 'Shivam', 'Raj'], 'Order_total':[10000, 20000, 30000, 40000] }

for name, total in zip(Customers['name'], Customers['Order_total']):
    if total > 20000:
        print(f'{name}: High value customer')
    else:
        print(f'{name}: Normal customer')
```

```python
def apply_discount(order_amount):
    discount = order_amount * 0.10
    final_price = order_amount - discount
    return final_price

order_amount = float(input('Enter the order amount: '))
final_price = apply_discount(order_amount)
print(f'Final price after discount: {final_price}')
```
