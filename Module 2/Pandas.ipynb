```python
# Fill Nulls with defaults
df["department"]= df["department"].fillna('Unknown')
df['salary']= df['salary'].fillna(df['salary'].median())
```

```python
display(df)
```

```python
#convert and Clearn Datae Column
df['join_date'] = pd.to_datetime(df['join_date'], errors='coerce')  # Convert invalid dates to NaT
df['join_date'] = df['join_date'].fillna(method='ffill')  # Fill NaT with forward fill
display(df)
```

# Why use dropna() vs fillna()?

✅ Answer:

- dropna() is used when missing data makes the row unusable.
- fillna() is used when a reasonable default (mean, median) can be applied.


# Real- World Messy to MasterData: Clean Real Data
- Real Data are messy and incomplete
- Today: Sales and customer data case studies
- Plus: Quick EDA using pandas_ profiling


## We have sales dataset[Messy Sales]
### - Sales.csv
order_id, customer_name, product, amount, date.
### - Common issues:
- Nulls in amount 
- Wrong date format 
- Extra whitespacess in product name 



```python
import pandas as pd 
df =pd.read_csv('sales_messy.csv')
display(df)
```

```python
display(df.head())
```

```python
display(df.tail())
```

```python
# Check the Null Count
print('Null Count: \n',df.isnull().sum())


```

```python
print('Data Type: \n',df.dtypes)
```

```python
print(df['product'])
```

```python
#Applying the function to remove extra spaces and camel case to product.
df["product"]= df["product"].str.strip().str.title()
display(df)

```

```python
print(df['product'].unique())
```

```python
# Handling the amount column - Filling the Missing values 
print(df['amount'].describe())
```

```python
# Fill the missing value with the median
df['amount']= df['amount'].fillna(df['amount'].median())
print(df)
```

```python
# Fix the date column - parse & clean invalid entries 
# error = 'coerce' is use to structure the date

df['date']= pd.to_datetime(df['date'],errors = 'coerce')
print('\n\n\nInvalid Date:',df['date'].isnull().sum())

#Fill Null Dates
df['date']= df['date'].fillna(method='ffill')

display(df)
```

```python
#Add some useful columns for Analystic like year and months

df['year']= df['date'].dt.year  
df['month']= df['date'].dt.month 
display(df)
```

# Final step is always do a sanity check.

```python
print(df.info())
print(df.describe())
print(df.head())
```
