## Spark Session is the entry Point from where it starts.

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("Spark DataFrames").getOrCreate()
print("SparkSession Created :", spark.version)
```

# Understand what is Pyspark Dataframe?

```python
#Create Spark Dataframe

data = [('Rahul',50000),('Shivam',40000),('Rajesh',45000),('Rakesh',55000)]
Column = ['Name','Salary']

spark_df = spark.createDataFrame(data, Column)
display(spark_df)
```

```python
# In spark the 
import pandas as pd 

pd_df = pd.DataFrame(data, columns = Column)
display(pd_df)
```

```python
spark_df.printSchema()
```

```python
print('Count in the Df', spark_df.count())
```

# Real-life Use Case.

These DataFrames are used for:

- Ingesting data from files
- Transforming at scale
- Writing to Delta Tables or Parquet files


```python
#Add Data Types 

from pyspark.sql.types import StructType,StructField, StringType, IntegerType

schema = StructType([
    StructField("Name", StringType(), True),
    StructField("Salary", IntegerType(), True)
])

df_typed = spark.createDataFrame(data, schema=schema)
df_typed.show()
df_typed.printSchema()
```
