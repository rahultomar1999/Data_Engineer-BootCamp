```python
from pyspark.sql import SparkSession 
spark = SparkSession.builder.appName("Spark DataFrames").getOrCreate()
```

```python
# Create Realistic  DataFrame for Employee
data = {('Rahul','DE',100000),
     ('shivam','HR',50000),
     ('Devraj','IT',1000000),
     ('Raj','BI',550000),
     ('Vishal','.Net',160000),
     ('Divya','Snow',170000),
     ('Prachi','java',150000),
}
     
Column = ['Name','Department','Salary']
df = spark.createDataFrame(data,Column)
display(df)
```

```python
# Select and apply Filter on DataFrame 
display(df.select('Name','Salary'))
```

```python
display(df.select('Salary','Name'))
```

```python
display(df.filter(df.Salary >100000))
```

```python
display(df.filter(df['Salary'] >100000))
```

```python
#Apply Multiple filter 
display(df.filter((df['Department']=='IT') & (df['Salary'] > 100000)))
```

```python
# Create s Derived Column 
from pyspark.sql.functions import col
#Add 10% Bonus to employee
df = df.withColumn('Bonus', col('Salary')* 0.1)
display(df)
```

```python
#GroupBY and Aggregation 

#Avg of Sslary by Department.

display(df.groupBy('Department').avg('Salary'))
```

```python
# Apply Multiple Aggregations

from pyspark.sql.functions import avg,max,min,sum,count

display(df.groupBy('Department').agg(avg('Salary').alias('Avg_Salary'),
                             max('Salary').alias('Max_Salary'),
                             min('Salary').alias('Min_Salary'),
                             sum('Salary').alias('Sum_Salary'),
                             count('Salary').alias('Count_Salary')))
```

```python
# Join() Combine two DataFrames

dept_data = [('IT', 101), ('HR', 102), ('DE', 103)]
dept_column = ['Department', 'DeptID']
dept_df = spark.createDataFrame(dept_data, dept_column)

#Join() Combine two DataFrames
joined_df = df.join(dept_df, on = 'Department', how = 'inner')
display(joined_df)


```

```python
# Transform V/s Action 
filtered  = df.filter(df.Salary > 100000)

#Action 
display(filtered)
```

## 📂 Build a DataFrame with the following schema: Name, Department, Salary, Location

```python

# Sample data
data = [
    ("Anurag", "IT", 70000, "Bangalore"),
    ("Priya", "HR", 55000, "Mumbai"),
    ("Ravi", "Finance", 65000, "Delhi"),
    ("Sneha", "IT", 60000, "Hyderabad")
]

columns = ["Name", "Department", "Salary", "Location"]

df_task = spark.createDataFrame(data, columns)

# Task 1: Filter IT employees with salary > 60K
df_task.filter((col("Department") == "IT") & (col("Salary") > 60000)).show()

# Task 2: Add column "Hike_Amount" as 15% of salary
df_task = df_task.withColumn("Hike_Amount", col("Salary") * 0.15)
df_task.show()

# Task 3: Group by Department and show average salary
df_task.groupBy("Department").avg("Salary").show()
```
